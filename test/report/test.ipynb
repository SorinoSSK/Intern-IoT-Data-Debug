```python
from get2 import *

# In[2]: get 3 days of data from unit 1
interval = get_seconds('3d')
unit_1 = Unit(70, 'Unit_1')
df = unit_1.get_values(70, interval)
df.head()
```

```python
timestamp = df['timestamp']
df.drop(labels=['timestamp'], axis=1, inplace=True)
df.insert(0, 'timestamp', timestamp)
df.head()
df.info()
```

###  timestamp index:
if using dt.weekday

0 : Monday

1 : Tuesday

2 : Wednesday

3 : Thursday

4 : Friday

5 : Saturday

6 : Sunday

```python
df.groupby([df['timestamp'].dt.date]).sum()
# df.head()
```

```python
bcu = df[['timestamp', 'bpow']]
bcu_sum_daily = bcu.groupby([bcu['timestamp'].dt.date]).sum()
bcu_mean_daily = bcu.groupby([bcu['timestamp'].dt.date]).mean()
bcu_sum_hourly = bcu.groupby([bcu['timestamp'].dt.date, bcu['timestamp'].dt.hour]).sum()
bcu_mean_hourly = bcu.groupby([bcu['timestamp'].dt.date, bcu['timestamp'].dt.hour]).mean()
```
